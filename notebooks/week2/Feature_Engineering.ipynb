```python
import pandas as pd

# Load the cleaned data we saved at the end of Week 1
df = pd.read_csv('../../data/earthquake_data.csv')

print("Cleaned data loaded successfully!")
```

```python
from sklearn.preprocessing import StandardScaler

# Normalize depth and magnitude as required
scaler = StandardScaler()
df[['magnitude_scaled', 'depth_scaled']] = scaler.fit_transform(df[['magnitude', 'depth']])

print("Data scaled!")
print(df[['magnitude', 'magnitude_scaled', 'depth', 'depth_scaled']].head())
```

```python
from sklearn.cluster import KMeans

# 1. Create clusters based on latitude and longitude
kmeans = KMeans(n_clusters=10, random_state=42, n_init=10)
df['geo_cluster'] = kmeans.fit_predict(df[['latitude', 'longitude']])

# 2. Add location risk score (average magnitude per cluster)
cluster_risk = df.groupby('geo_cluster')['magnitude'].mean().reset_index()
cluster_risk.rename(columns={'magnitude': 'location_risk_score'}, inplace=True)

df = df.merge(cluster_risk, on='geo_cluster', how='left')
print("Clusters and Risk Scores created!")
print(df[['geo_cluster', 'location_risk_score']].head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode categorical columns if they exist (like 'alert' level)
if 'alert' in df.columns:
    encoder = LabelEncoder()
    df['alert'] = df['alert'].fillna('unknown')
    df['alert_encoded'] = encoder.fit_transform(df['alert'])
    print("Alert column encoded!")

# Save the final, engineered dataset for Week 3 (Model Training)
df.to_csv('../../data/engineered_earthquake_data.csv', index=False)
print("Engineered dataset saved for model training!")
```
