```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium

pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")

print("Libraries Imported Successfully")
```

```python
df = pd.read_csv(r"C:\Users\mssha\OneDrive\Desktop\ImpactSense\data\earthquake dataset.csv")
print("Dataset Loaded Successfully")
df.head()
```

```python
print("Dataset Shape:", df.shape)

print("\nColumn Names:")
print(df.columns)
```

```python
df.info()
```

```python
print("\nMissing Values:")
df.isnull().sum()
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['magnitude'], bins=30, kde=True)

plt.title("Magnitude Distribution")
plt.xlabel("Magnitude")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['depth'], bins=30, kde=True)

plt.title("Depth Distribution")
plt.xlabel("Depth")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['sig'], bins=30)

plt.title("Earthquake Significance")
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x=df['depth'],
    y=df['magnitude'],
    hue=df['alert'],     # different colors
    palette="Set1"
)

plt.title("Depth vs Magnitude by Alert Level")
plt.xlabel("Depth")
plt.ylabel("Magnitude")

plt.legend(title="Alert Level")
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)

plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x=df['alert'])

plt.title("Earthquake Alert Levels")
plt.show()
```

```python
df.to_csv("earthquake_week1_cleaned.csv",
          index=False)

print("Week 1 dataset saved successfully!")
```
