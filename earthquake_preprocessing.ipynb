# Earthquake Data Preprocessing & Feature Engineering
This notebook focuses on cleaning the dataset, creating new features, and normalizing numerical values for machine learning.

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

## 2. Load Dataset and Initial Cleaning

```python
file_path = "earthquake_alert_balanced_dataset.csv"
df = pd.read_csv(file_path)

# Drop duplicates
initial_shape = df.shape
df = df.drop_duplicates()
print(f"Dropped {initial_shape[0] - df.shape[0]} duplicate rows.")
print(f"Current shape: {df.shape}")
```

## 3. Handle Missing Values

```python
# Check for nulls
null_counts = df.isnull().sum()
print("Null values per column:\n", null_counts)
```

## 4. Feature Engineering

```python
# 4.1 Create a new feature: depth_magnitude_ratio
df['depth_mag_ratio'] = df['depth'] / (df['magnitude'] + 1)

# 4.2 Encode the categorical 'alert' column
label_encoder = LabelEncoder()
df['alert_encoded'] = label_encoder.fit_transform(df['alert'])

print("New features created. First 5 rows:")
display(df.head())
print("\nAlert mapping:", dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))
```

## 5. Normalize Numerical Features

```python
# Select numerical columns to scale
numerical_features = ['magnitude', 'depth', 'cdi', 'mmi', 'sig', 'depth_mag_ratio']

scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])

print("Normalization complete. First 5 rows of scaled features:")
display(df[numerical_features].head())
```

## 6. Save Preprocessed Data

```python
output_path = "earthquake_processed.csv"
df.to_csv(output_path, index=False)
print(f"Preprocessed data saved to {output_path}")
```
